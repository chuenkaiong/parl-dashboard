```python
def get_hansard_report(date):
    datestr = date.strftime('%d-%m-%Y')
    url = 'https://sprs.parl.gov.sg/search/getHansardReport/?sittingDate=' + datestr
    r = requests.post(url)
    if r.status_code == 200:
        return r.json()
    

def get_sitting_dates(stop_date):
    for i in range(10):
        url = f'https://www.sgpc.gov.sg/page-{i}?agency=PARL'
        r = requests.get(url)
        if r.status_code == 200:
            soup = BeautifulSoup(r.text)
            dates = []
            for div in soup('div', class_='press-title'):
                title = div.get_text().strip().upper()
                if 'SITTING ON' in title:
                    datestr = title.split('SITTING ON')[-1]
                    date = parser.parse(datestr, ignoretz=True).date()
                    if date == stop_date:
                        break
                    if date not in dates:
                        dates.append(date)
            return dates
        
def clean_name(name):
    prefixes = ['Assoc', 'Asst', 'Dr', 'Er', 'Mdm', 'Miss', 'Mr', 'Mrs', 'Ms', 'Prof']
    first, rest = name.split(maxsplit=1)
    if first.strip() in prefixes:
        return clean_name(rest)
    return name
```

```python
def get_historic_sitting_dates():
    url = 'https://www.parliament.gov.sg/parliamentary-business/votes-and-proceedings?parliament=&fromDate=&toDate=&page='
    dates = dict()
    for i in range(1, 44):
        page_url = url + str(i) + '&pageSize=10'
        r = requests.get(page_url)
        if r.status_code != 200:
            print('Failed at: ' + page_url)
            continue
        soup = BeautifulSoup(r.text)
        for div_wrap in soup('div', class_='votes-proceedings-wrap'):
            boxgaps = div_wrap('div', class_='col-md-3 col-xs-12 xs-boxgap')
            for i in range(0, len(boxgaps), 2):
                div_sitting_date, div_parl = boxgaps[i:i+2]
                datestr = div_sitting_date.get_text().split('on ')[-1]
                date = parser.parse(datestr, ignoretz=True).date()
                parlstr = div_parl.get_text().split()[0].strip('th')
                dates[parlstr] = dates.get(parlstr, []) + [date]
    return dates
```

```python
historic_dates = get_historic_sitting_dates()
```

```python
mp_attendances = dict()
for report in hansard_13th[::-1]:
    for mp in report['attendanceList']:
        if 'www.' in mp['mpName']:
            continue
        name = mp['mpName'].strip().split(', ')[0].rsplit(' (', maxsplit=1)[0]
        attendance = mp['attendance']
        if not mp_attendances.get(name):
            mp_attendances[name] = {'present': 0, 'absent': 0}
            if 'Nominated Member' in mp['mpName']:
                mp_attendances[name]['NMP'] = True
            elif 'Non-Constituency Member' in mp['mpName']:
                mp_attendances[name]['NCMP'] = True
        if attendance:
            mp_attendances[name]['present'] += 1
        else:
            mp_attendances[name]['absent'] += 1
```
